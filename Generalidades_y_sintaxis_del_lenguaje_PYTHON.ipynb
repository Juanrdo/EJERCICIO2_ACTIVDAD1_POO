**Interpretado**

<div align="justify">

Los lenguajes interpretados son aquellos en los que el c√≥digo se traduce y ejecuta mediante un int√©rprete a medida que se necesita, en lugar de ser compilado previamente. Algunos ejemplos comunes de lenguajes interpretados son Python, Ruby y JavaScript.

- Facilita la colaboraci√≥n: La simplicidad y legibilidad de Python permiten que varios desarrolladores trabajen en el mismo proyecto sin grandes barreras de comprensi√≥n, lo que reduce los costos de mantenimiento y mejora la eficiencia del equipo.

**Tipado din√°mico**

<div align="justify">

El tipado din√°mico significa que en un lenguaje de programaci√≥n como Python, las variables pueden cambiar su tipo de datos durante la ejecuci√≥n del programa. No es necesario definir de antemano el tipo de dato que va a contener la variable, ya que Python determina autom√°ticamente el tipo seg√∫n el valor que se le asigne.


</div>

**üßë‚ÄçüíªEjemplo 1üë©‚Äçüíª**

```python

x = 10        # x es un entero (int)
x = "Hola"    # Ahora x es una cadena de texto (str)
x = 3.14      # Ahora x es un n√∫mero decimal (float)
```

<center>

**üßë‚ÄçüíªEjemplo 2üë©‚Äçüíª**

```python

a = 1
print(type(a))
# En este punto, 'a' tiene el valor 1, que es un n√∫mero entero (int).
# Por lo tanto, type(a) devolver√° <class 'int'>.

a = "b"
print(type(a))
# Ahora 'a' se ha reasignado a la cadena de texto "b".
# Python cambia autom√°ticamente el tipo de 'a' a 'str' (cadena de texto).
# type(a) devolver√° <class 'str'>.

a = []
print(type(a))
# Finalmente, 'a' se reasigna a una lista vac√≠a ([]).
# Una vez m√°s, Python ajusta autom√°ticamente el tipo de 'a', que ahora es 'list'.
# type(a) devolver√° <class 'list'>.
```

**Conversi√≥n de tipos**

<div align="justify">

La conversi√≥n de datos o cast se refiere al proceso de cambiar un valor de un tipo de dato a otro. Existen dos tipos principales de conversi√≥n de datos: conversi√≥n impl√≠cita y conversi√≥n expl√≠cita, cada una utilizada en diferentes situaciones dependiendo de si el lenguaje permite la conversi√≥n autom√°tica o requiere intervenci√≥n manual.

</div>

- **Conversi√≥n impl√≠cita**

<div align="justify">

La conversi√≥n impl√≠cita de tipos en Python ocurre de manera autom√°tica, sin que el programador tenga que intervenir. Python realiza esta conversi√≥n cuando se llevan a cabo operaciones entre dos tipos de datos diferentes. Aunque suele ser √∫til, es importante entender c√≥mo funciona para evitar errores o resultados inesperados en el programa.


</div>

**üßë‚ÄçüíªEjemplo 13üë©‚Äçüíª**

```python

a = 1  # Variable 'a' es un entero (int)
b = 2.3  # Variable 'b' es un decimal (float)

a = a + b  # Python convierte autom√°ticamente 'a' de int a float
print(type(a))  # Resultado: <class 'float'>
# La conversi√≥n impl√≠cita ocurre aqu√≠ porque al sumar un entero y un decimal, Python autom√°ticamente convierte el resultado a float.

# Sin embargo, hay casos en los que no se puede realizar la conversi√≥n autom√°ticamente
a = 1  # Variable 'a' es un entero (int)
b = "2.3"  # Variable 'b' es una cadena de texto (string)

c = a + b  # Este c√≥digo genera un error de tipo (TypeError)
# Python no puede sumar un entero con una cadena directamente, por lo que se debe hacer una conversi√≥n expl√≠cita.
```

- **Conversi√≥n explicita**

<div align="justify">

En Python, tambi√©n podemos realizar conversiones expl√≠citas entre tipos de datos utilizando funciones espec√≠ficas que proporciona el lenguaje. Estas conversiones, tambi√©n conocidas como castings, nos permiten cambiar manualmente el tipo de un valor a otro. A continuaci√≥n se presentan las funciones de conversi√≥n m√°s utilizadas:


</div>

- **int() ‚Äì Convierte a entero:**

     - Convierte un valor num√©rico o una cadena que representa un n√∫mero a tipo entero (int).

**üßë‚ÄçüíªEjemplo 14üë©‚Äçüíª**

```python

a = int(3.5)  # Convierte el n√∫mero decimal 3.5 a 3 (entero)
b = int("10")  # Convierte la cadena "10" en el entero 10
print(a, b)  # Imprime: 3 10
```

- **float() ‚Äì Convierte a decimal:**

     - Convierte un valor num√©rico o una cadena que representa un n√∫mero a tipo decimal (float).

**üßë‚ÄçüíªEjemplo 15üë©‚Äçüíª**

```python

a = float(5)  # Convierte el entero 5 a 5.0 (decimal)
b = float("2.3")  # Convierte la cadena "2.3" a 2.3 (decimal)
print(a, b)  # Imprime: 5.0 2.3
```


- **str() ‚Äì Convierte a cadena de texto:**

     - Convierte un valor num√©rico o cualquier otro tipo de dato a tipo cadena (str).

**üßë‚ÄçüíªEjemplo 16üë©‚Äçüíª**

```python

a = str(10)  # Convierte el entero 10 a la cadena "10"
b = str(3.14)  # Convierte el decimal 3.14 a la cadena "3.14"
print(a, b)  # Imprime: 10 3.14
```

- **bool() ‚Äì Convierte a booleano:**

     - Convierte un valor a tipo booleano (bool), donde 0, None, y cadenas vac√≠as son consideradas False, mientras que otros valores son considerados True

**üßë‚ÄçüíªEjemplo 17üë©‚Äçüíª**

```python

a = bool(1)  # Convierte el valor 1 a True
b = bool(0)  # Convierte el valor 0 a False
c = bool("")  # Cadena vac√≠a se convierte a False
print(a, b, c)  # Imprime: True False False
```

**Estructuras condicionales y bucles en Python**

- **if (condicional)**

    - La estructura if permite ejecutar un bloque de c√≥digo solo si una condici√≥n es verdadera.

**üßë‚ÄçüíªEjemplo 22üë©‚Äçüíª**

```python

edad = 18

if edad >= 18:
    print("Eres mayor de edad.")
```

- **if - else (condicional con alternativa)**
    - Si la condici√≥n del if es falsa, puedes usar el bloque else para ejecutar un bloque de c√≥digo alternativo.

**üßë‚ÄçüíªEjemplo 23üë©‚Äçüíª**

```python

edad = 16

if edad >= 18:
    print("Eres mayor de edad.")
else:
    print("Eres menor de edad.")
```

- **if - elif - else (m√∫ltiples condiciones)**

    - Cuando tienes varias condiciones, puedes usar elif para agregar m√°s comprobaciones.

**üßë‚ÄçüíªEjemplo 24üë©‚Äçüíª**

```python

nota = 85

if nota >= 90:
    print("Excelente")
elif nota >= 70:
    print("Aprobado")
else:
    print("Reprobado")
```

- **for (bucle iterativo)**

   - El bucle for en Python se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario, cadena o rango).

**üßë‚ÄçüíªEjemplo 25üë©‚Äçüíª**

```python

frutas = ["manzana", "banana", "naranja"]

for fruta in frutas:
    print(fruta)
```

- **for-else (bucle con clausula else)**

    - El bucle for-else es una caracter√≠stica √∫nica de Python. El bloque else se ejecuta cuando el bucle termina normalmente, es decir, si no se interrumpe con un break.

**üßë‚ÄçüíªEjemplo 26üë©‚Äçüíª**

```python

numeros = [1, 2, 3, 4]

for num in numeros:
    if num == 5:
        print("Encontrado 5")
        break
else:
    print("No se encontr√≥ el n√∫mero 5")
```

- **while (bucle condicional)**

    - El bucle while repite el bloque de c√≥digo mientras una condici√≥n sea verdadera.

**üßë‚ÄçüíªEjemplo 27üë©‚Äçüíª**

```python

contador = 0

while contador < 5:
    print(contador)
    contador += 1
```

- **while-else (bucle con clausula else)**

    - Al igual que con for-else, puedes usar else con un bucle while. El bloque else se ejecuta si el bucle no es interrumpido por un break.

**üßë‚ÄçüíªEjemplo 28üë©‚Äçüíª**

```python

contador = 0

while contador < 5:
    print(contador)
    contador += 1
else:
    print("El contador ha alcanzado 5.")
```
